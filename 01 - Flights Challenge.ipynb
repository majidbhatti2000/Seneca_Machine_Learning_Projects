#### Finding Outliers

```python
from matplotlib import pyplot as plt

# Create a function that we can re-use
# call as show_distribution(df_students['Grade'])
def show_distribution(var_data, label=""):
    '''
    This function will make a distribution (graph) and display it
    '''

    # Get statistics
    min_val = var_data.min()
    max_val = var_data.max()
    mean_val = var_data.mean()
    med_val = var_data.median()
    mod_val = var_data.mode()[0]

    print('Minimum:{:.2f}\nMean:{:.2f}\nMedian:{:.2f}\nMode:{:.2f}\nMaximum:{:.2f}\n'.format(min_val,
                                                                                            mean_val,
                                                                                            med_val,
                                                                                            mod_val,
                                                                                            max_val))

    # Create a figure for 2 subplots (2 rows, 1 column)
    fig, ax = plt.subplots(2, 1, figsize = (10,4))

    # Plot the histogram   
    ax[0].hist(var_data)
    ax[0].set_ylabel('Frequency')

    # Add lines for the mean, median, and mode
    ax[0].axvline(x=min_val, color = 'gray', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=mean_val, color = 'cyan', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=med_val, color = 'red', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=mod_val, color = 'yellow', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=max_val, color = 'gray', linestyle='dashed', linewidth = 2)

    # Plot the boxplot   
    ax[1].boxplot(var_data, vert=False)
    ax[1].set_xlabel('Value')

    # Add a title to the Figure
    fig.suptitle(i + ' - Data Distribution')

    # Show the figure
    #fig.show()

delayVar = ('ArrDelay', 'DepDelay')
for i in delayVar:
    show_distribution(df_flights[i], i)
```

#### Eliminating Outlier

```python
# Trim outliers for ArrDelay based on 1% and 90% percentiles
print("Total Rows in dataset before trim:", df_flights.count)
arrdly_001 = df_flights.ArrDelay.quantile(0.01)
arrdly_090 = df_flights.ArrDelay.quantile(0.90)
df_flights = df_flights[df_flights['ArrDelay'] > arrdly_001]
df_flights = df_flights[df_flights['ArrDelay'] < arrdly_090]
print("Total Rows in dataset after trim:", df_flights.count)
```

```python
# Trim outliers for DepDelay based on 1% and 90% percentiles
print("Total Rows in dataset before trim:", df_flights.count)
depdly_001 = df_flights.DepDelay.quantile(0.01)
depdly_090 = df_flights.DepDelay.quantile(0.90)
df_flights = df_flights[df_flights['DepDelay'] > depdly_001]
df_flights = df_flights[df_flights['DepDelay'] < depdly_090]
print("Total Rows in dataset after trim:", df_flights.count)
```

```python
for i in delayVar:
    show_distribution(df_flights[i], i)
```

## Explore the data

```python
df_flights.describe()
```

##### What are the average (mean) departure and arrival delays?

```python
delayFields = ['DepDelay','ArrDelay']

df_flights[delayFields].mean()
```

##### How do the carriers compare in terms of arrival delay performance?

```python
for col in delayFields:
    df_flights.boxplot(column=col, by='Carrier')
```

##### Are some days of the week more prone to arrival days than others?

```python
for col in delayFields:
    df_flights.boxplot(column=col, by='DayOfWeek')
```

Which departure airport has the highest average departure delay?

```python
df_highDelay = df_flights.groupby('OriginAirportName')
df_highDelay = pd.DataFrame(df_highDelay['DepDelay'].mean()).sort_values("DepDelay", ascending=False)
df_highDelay
df_highDelay.plot(kind='bar', figsize=(12,12))
#mean_departure_delays = pd.DataFrame(departure_airport_group['DepDelay'].mean()).sort_values('DepDelay', ascending=False)
```

Do late* departures tend to result in longer arrival delays than on-time departures?*

```python
df_flights.boxplot(column='ArrDelay', by='DepDel15', figsize=(6,6))
```

Which route (from origin airport to destination airport) has the most late arrivals?

```python
route = pd.Series(df_flights['OriginAirportName'] + ' > ' + df_flights['DestAirportName'])
df_flights = pd.concat([df_flights,route.rename('Route')],axis=1)
df_flights.head()
```

```python
route_group = df_flights.groupby('Route')
pd.DataFrame(route_group['ArrDel15'].sum()).sort_values("ArrDel15",ascending=False)
```

Which route has the highest average arrival delay?

```python
pd.DataFrame(route_group['ArrDelay'].sum()).sort_values("ArrDelay",ascending=False)
```
